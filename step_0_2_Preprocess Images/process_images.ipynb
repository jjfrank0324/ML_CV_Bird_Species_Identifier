# Find minimum size of images

```python
min_x = 1000
min_y = 1000
```

```python
cnt = 0
for folder in image_folders:
    if folder == ".DS_Store":
        continue
    
    os.chdir(os.path.join(path, images_dir, folder))
    images = os.listdir()
    for i in images:

        print("processing image ", cnt)
        
        image = cv2.imread(i)
        y, x = image.shape[:2]
        
        if x < y:  # need rotate 
            image = cv2.rotate(image, cv2.cv2.ROTATE_90_CLOCKWISE)
            x, y = y, x
        
        min_x = min(min_x, x)
        min_y = min(min_y, y)

        # save image 
        cv2.imwrite(os.path.join(path, processed_image_dir, folder, i), image)
        
        cnt += 1
```

# Resize all images to the minimum size

```python
# we could get 
min_x = 264
min_y = 121
```

```python
cnt = 0
for folder in image_folders:
    if folder == ".DS_Store":
        continue
    
    os.chdir(os.path.join(path, processed_image_dir, folder))
    images = os.listdir()
    for i in images:

        print("processing image ", cnt)
        
        image = cv2.imread(i)

        # resize image
        image = cv2.resize(image, dsize=(min_x, min_y))

        # save image 
        cv2.imwrite(os.path.join(path, processed_image_dir, folder, i), image)
        
        cnt += 1
```
