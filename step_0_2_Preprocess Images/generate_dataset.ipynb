```python
def get_id_attr_relationship(file="attributes.txt"):
    """
    Given a path to the attributes file, returns a dict maps attributes id to attributes name 
    """
    id_to_attr = dict()

    with open(file, "r") as f:
        for line in f:
            id_and_attr = line.split(" ")
            id = int(id_and_attr[0])
            attr = id_and_attr[1].strip()
            id_to_attr[id] = attr

    return id_to_attr
```

```python
def get_folder_to_bird(image_folders):
    folder_to_bird = {}
    for folder_name in image_folders:
        bird_name = folder_name.split(".")[1].replace("_", " ")
        folder_to_bird[folder_name] = bird_name
    return folder_to_bird
```

```python
def get_image_path_to_attr_df(id_to_attr, image_folders, folder_to_bird, image_dir):
    df = pd.DataFrame(columns=["image_path"] + [id_to_attr[i] for i in range(288)])

    idx = 0
    for folder in image_folders:
        if folder.startswith('.'):
            continue
        os.chdir(os.path.join(path, image_dir, folder))
        images = os.listdir(os.getcwd())
        
        bird_name = folder_to_bird[folder]
        attrs = bird_to_attrs_df.loc[bird_to_attrs_df['bird_name'] == bird_name].iloc[:, 1:]

        for image_name in images:
            if ("(1)" in image_name):  # duplicates
                continue
            if (image_name.startswith('.')):
                continue
            print("processing image ", idx)
            df.loc[idx] = [os.path.join(image_dir, folder, image_name)] + attrs.values.tolist()[0]
            idx += 1
    
    return df
```

```python
id_to_attr = get_id_attr_relationship("C:\\Users\\czahi\\Desktop\\CIS519-Final-Project\\step_0_2_Preprocess Images\\attributes.txt")
folder_to_bird = get_folder_to_bird(image_folders)
df = get_image_path_to_attr_df(id_to_attr, image_folders, folder_to_bird, image_dir)
```

```python
os.chdir(os.path.join(path, "data\\csv"))
df.to_csv('seen_images.csv', index=False)
```

# Dataset for Pytorch CNN

```python
import torch
```

```python
class CUBDataset(Dataset):
    def __init__(self, csv_file, root_dir, transform=None):
        self.df = pd.read_csv(csv_file)

        # for test, only first 200 samples
#         self.df = self.df.head(200)

        self.transform = transform
        self.root_dir = root_dir
        
    def __len__(self):
        return len(self.df)    
    
    def __getitem__(self, idx):
        row = self.df.loc[idx]
        img_path, img_label = row['image_path'], torch.tensor(row[1:].astype('float32').values)
        img_path = os.path.join(self.root_dir, img_path)
        img = Image.open(img_path)
        if self.transform:
            img = self.transform(img)
        return img, img_label
```

```python
transform = transforms.Compose([transforms.ToTensor()])
dataset = CUBDataset(os.path.join(path, "data\\csv\\seen_images.csv"), path, transform=transform)
```
